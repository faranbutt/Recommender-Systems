```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd
```

```python
data = {
     'title': ['Inception', 'The Matrix', 'Interstellar', 'The Dark Knight'],
    'features': [
        'Sci-Fi Action Christopher Nolan Leonardo DiCaprio',
        'Sci-Fi Action Wachowskis Keanu Reeves',
        'Sci-Fi Drama Christopher Nolan Matthew McConaughey',
        'Action Crime Christopher Nolan Christian Bale'
    ]
}
df = pd.DataFrame(data)
df
```

```python
vectorizer = TfidfVectorizer()
tfdi_Matrix = vectorizer.fit_transform(df['features'])
```

```python
cosine_sim = cosine_similarity(tfdi_Matrix)
cosine_sim
```

```python
def recommender(movie_title,df,cosine_sim):
  print("Faran")
  idx = df.index[df['title'] == movie_title][0]
  sim_scores = list(enumerate(cosine_sim[idx]))
  sim_scores = sorted(sim_scores,key=lambda x:x[1],reverse=True)
  print(sim_scores)
  sim_scores = sim_scores[1:3]
  movie_indices = [i[0] for i in sim_scores]
  return df['title'].iloc[movie_indices]
```

```python
recommender("Inception",df,cosine_sim)
```
